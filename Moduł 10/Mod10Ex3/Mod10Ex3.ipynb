Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij Mentorowi jako Jupyter Noteboook zapisany w Github.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000

```python
mean = 3
std = 1
snd = stats.norm(mean, std)

x = np.linspace(0,10,100)
y = np.linspace(0,10,10000)
```

```python
plt.figure(figsize=(7.5,7.5))
plt.plot(x, snd.pdf(x))
plt.title(f'Rozkład normarny (Średnia = {mean}, STD = {std})', fontsize='15')
plt.xlabel('Wartości zmiennej losowej x', fontsize='15')
plt.ylabel('Prawdopodobieństwo', fontsize='15')
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.plot(y, snd.pdf(y))
plt.title(f'Rozkład normarny (Średnia = {mean}, STD = {std})', fontsize='15')
plt.xlabel('Wartości zmiennej losowej x', fontsize='15')
plt.ylabel('Prawdopodobieństwo', fontsize='15')
plt.show()
```

 Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej.

Dla zbioru do 100 (x)

```python
mean_x = np.round(np.mean(x))
print(f"Średnia x: {mean_x}")

median_x = np.median(x)
print(f"Mediana x: {median_x}")

mode_x = stats.mode(x, keepdims=True)
print(f"Dominanta x: {mode_x}")

x_q0 = np.quantile(x, 0.0)
x_q1 = np.round(np.quantile(x, 0.25))
x_q3 = np.round(np.quantile(x, 0.75))
x_q4 = np.round(np.quantile(x, 1.0))
print(f"Q0 dla x: {x_q0}")
print(f"Q1 dla x: {x_q1}")
print(f"Q2 dla x: {median_x}")
print(f"Q3 dla x: {x_q3}")
print(f"Q4 dla x: {x_q4}")

range_x = max(x) - min(x)
print(f"Zakres dla x: {range_x}")

iqr_x = x_q3 - x_q1
print(f"IQR(x): {iqr_x}")

variance_x = np.var(x, ddof=1)
print(f"Wariancja x: {variance_x}")

standard_deviation_x = np.std(x, ddof=1)
print(f"Odchylenie standardowe x: {standard_deviation_x}")
```

Dla zbioru do 1000 (y)

```python
mean_y = np.round(np.mean(y))
print(f"Średnia y: {mean_y}")

median_y = np.median(y)
print(f"Mediana y: {median_y}")

mode_y = stats.mode(y, keepdims=True)
print(f"Dominanta y: {mode_y}")

y_q0 = np.quantile(y, 0.0)
y_q1 = np.round(np.quantile(y, 0.25))
y_q3 = np.round(np.quantile(y, 0.75))
y_q4 = np.round(np.quantile(y, 1.0))
print(f"Q0 dla y: {y_q0}")
print(f"Q1 dla y: {y_q1}")
print(f"Q2 dla y: {median_y}")
print(f"Q3 dla y: {y_q3}")
print(f"Q4 dla y: {y_q4}")

range_y = max(y) - min(y)
print(f"Zakres dla x: {range_y}")

iqr_y = y_q3 - y_q1
print(f"IQR(y): {iqr_y}")

variance_y = np.var(y, ddof=1)
print(f"Wariancja y: {variance_y}")

standard_deviation_y = np.std(y, ddof=1)
print(f"Odchylenie standardowe y: {standard_deviation_y}")
```
